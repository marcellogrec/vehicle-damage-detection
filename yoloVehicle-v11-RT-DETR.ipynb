```python
# 1 Criar o treinamento 11S

from ultralytics import YOLO

#model = YOLO("yolov10n.yaml")  # build a new model from YAML
model = YOLO(f'{HOME}/weights/yolov11s.pt')  # load a pretrained model (recommended for training)
#model = YOLO(f'{HOME}/weights/yolov10b.pt')  # load a pretrained model (recommended for training)
#model = YOLO(f'{HOME}/weights/yolov10m.pt')  # load a pretrained model (recommended for training)
#model = YOLOv10("yolov10n.yaml").load(f'{HOME}/weights/yolov10n.pt')  # build from YAML and transfer weights

# Train the model
model.train(data= f'{HOME}/datasets/avaria/data_config.yaml',
            epochs=100,
            device='0',
            imgsz=640,
            cache=False,
            batch=-1)

model.export(format="onnx")
```

```python
# 2 Criar o treinamento 11N

from ultralytics import YOLO

model = YOLO(f'{HOME}/weights/yolov11n.pt')

model.train(data= f'{HOME}/datasets/avaria/data_config.yaml',
            epochs=100,
            device='0',
            imgsz=640,
            cache=False,
            batch=-1)

model.export(format="onnx")
```
